## 1. When $X$ influence $Y$ ?

$X \rightarrow Y$, $X$ influences $Y$, 直接影响               
$X \leftarrow Y$, Condition on $X$ changes beliefs in $Y$， 给定$X$影响$Y$                
$X \rightarrow W \rightarrow Y$, indirect influcnece of $X$ on $Y$ through $Y$， 通过$W$间接影响       
$X \leftarrow W \leftarrow Y$, Condition $Y$ changes $W$ then $X$， 反着来    
$X \leftarrow W \rightarrow Y$, Condition $Y$ changes $W$ then $X$， 中间向两边的   
$X \rightarrow W \leftarrow Y$ V structure is no an active trial <font color='darkred'>XXX</font>  这个是不对的   


**A TRIAL $X_{1} - X_{2} - ... - X_{n}$ (can be either direction) is active if it contains no V structure $X_{i-1} \rightarrow X_{i} \leftarrow X_{i+1}$** 不含V字型的线索，是一条活的线索

## 2. It is always helperful to learn bayesian statistics with exmaples
###  什么情况下$X$会影响$Y$
$X$ = 你穷 $\rightarrow$ $Y$ = 你借钱                         
1. $X$会影响$Y$， 你穷了，你才去借钱
2. 当我知道你借钱，given $Y$ = 借钱， 我觉得你穷的概率很大。 $X \rightarrow Y$这个关系是存在的，但当我们给定$Y$的时候，我们会改变对X的看法。    
3. $X$ = 你穷 $\rightarrow$ $Y$ = 你借钱 $\rightarrow$  $Z$ = 你被催收打了, 显然，你穷了（钱不够了）, 你被催收打的概率肯定大.    
4. 当我知道你被催收打了的时候，我知道你借钱了，或者你朋友/家人，借钱了，那我觉得你就一个穷屌丝。 $Z -influence -Y -influence -X$   
5. $X = 你缺乏营养 \leftarrow W = 你穷 \rightarrow Y你借钱$, 当我知道你缺乏营养的时候，我感觉你穷的概率变大了，当我感觉你穷的时候，你借钱的概率变大了。    
6. $X = 你爸妈没钱 \rightarrow W = 你穷 \leftarrow Y你学习差$， 当我知道你穷的时候， 你爸妈没钱和你学习差就没啥关系了， 所以这里吧，X影响不了Y

## Active Trial
一个序列 $x_{1}, x_{2}, ... , x_{n}$                        
这个序列中所有的V结构$x_{i-1} \rightarrow x_{i} \leftarrow x_{i+1}$ 中， $x_{i}$或者至少有那么一个子孙，被观测到了。               
其他的不再V结构的节点，都没有被观测到。
